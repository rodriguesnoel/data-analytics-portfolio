<h2><font color=#a12a04>release_date vs year</font></h2>

```python
# Data cleaning
# As there is no difference between 'release_date' and 'year', we will drop 'release_date' as it does not serve any significant purpose.

data.drop('release_date', axis=1, inplace=True)
```

```python
# Check if 'release_date' column has been dropped

data.info()
```

<h2><font color=#a12a04>duplicate song and name</font></h2>

```python
# Data cleaning
# Let's find if there are any duplicate songs with same 'artists' and 'name'

data['artists+name'] = data.apply(lambda row: row['artists'] + row['name'], axis=1)
```

```python
df = data[data['artists+name'].duplicated()]
```

```python
df.shape
```

```python
# Omg, there are 14,948 duplicates!!!
```

```python
# For example, there are three 'The Most Beautiful Girl' by Charlie Rich.
# The 'popularity' ranges from 24 to 57 is calculated by Spotify's algorithm based on the total number of plays.
# Because the same song can be released multiple times, they are therefore rated independently. 

data[data['name']=='The Most Beautiful Girl']
```

```python
# We therefore decided to only keep the duplicated song with the highest 'popularity' rating and drop the rest.
# We gather the list of duplicates corresponding to the maximum popularity for each duplicated artist/song pair.

max_dup_list = []

for name in df['artists+name'].unique():
    subset = data[data['artists+name'] == name].copy()
    m = subset['popularity'].max()
    index = subset[subset['popularity'] == m].index[0]
    max_dup_list.append(index)
```

```python
# We have now removed all duplicated songs and kept one remaining song with the highest 'popularity' rating
# This code checks if there are any more duplicates. Answer: No

check_anymore_dup = data.loc[max_dup_list].copy()
check_anymore_dup['artists+name'].duplicated().sum()
```

```python
# Now let's loop through each row to see if there are any further duplicates. And if so, remove them.

dup_loop = data.copy()
for i, row in data.iterrows():
    if row['artists+name'] in df['artists+name'].unique():
        dup_loop.drop(index=i, inplace=True)
```

```python
# Concatenate all the removed duplicates

all_dup = [check_anymore_dup, dup_loop]
no_dup_data = pd.concat(all_dup)
```

```python
# Let's check if anymore duplicates

no_dup_data['artists+name'].duplicated().sum()
```

```python
# Now, we can see that all the duplicated rows have been deleted. Previously, there was 174,389 rows (14,948 dropped).

no_dup_data.shape
```

<h2><font color=#a12a04>remove id and artists+name<font></h2>

```python
# We can now remove the 'artists+name' column as it was only meant to look for duplicates.
# We can also remove the 'id' column as it's incomprehensible.

no_dup_data.drop(['artists+name', 'id'], axis=1, inplace=True)
```

```python
# We now have 159,441 rows and 17 columns

no_dup_data.shape
```

```python
# Let's give a new name to 'no_dup_data' for easy further reference

new_data = no_dup_data
```

```python
new_data.head()
```

<h2><font color=#a12a04>remove zero values from tempo</font></h2>

```python
# Check for any 'tempo' row with zero value and remove those rows

new_data = new_data[new_data['tempo'] != 0].copy()
```

```python
new_data.reset_index(inplace=True, drop=True)
```

```python
new_data.shape
```

<h2><font color=#a12a04>energy vs loudness / remove loudness<font></h2>

```python
# In the Spotify API description, it says that `energy` measures the intensity and activity (ie. fast, loud and noisy)
# On the other hand, 'loudness' measures the quality and physical strength of amplitutde (ie. volume).
# Let's check if energy and loudness are strongly correlated. If so, there is "multicollinearity" (which is bad).

m = new_data['energy'].corr(new_data['loudness'])
print(f'The Pearson correlation coefficient is: {m:^10}')
```

```python
# Another way to check is using a heatmap

plt.subplots(figsize=(24, 12))
sns.heatmap(new_data.corr(), annot=True, square=True)
plt.show()
```

```python
# With a correlation coefficient of 0.777 (near to 1), we can arguably drop 'loudness' to prevent multicolinearity.

new_data.drop('loudness', axis=1, inplace=True)
```

```python
new_data.shape
```

<h2><font color=#a12a04>remove zero values from popularity<font></h2>

```python
# If we take a look at 'popularity', we see that there are zero values. So let's remove them.

song = new_data[new_data['popularity'] != 0].copy()
song.reset_index(inplace=True, drop=True)
```

<h2>4.3 Data Preparation</h2>
<br>
First, we will divide our independent and dependent variable into two separate variables.
Second, we will split the data into training and testing datasets.

```python
# Import necessary libraries for testing, training and modelling

from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, RobustScaler, OneHotEncoder
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split, RandomizedSearchCV
```

```python
# First, we will divide our independent and dependent variable by dropping 'popularity' (i.e. independent variable).
# By sorting them by 'year'

X = song.sort_values(by='year').drop(columns=['popularity']).copy()
y = song.sort_values(by='year')['popularity'].copy()
```

```python
# Second, we will split the data into training and testing datasets.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle=False)
```

```python
X_train.tail()
```

```python
X_val.tail()
```

<h2>4.4 Training the Model</h2>
<br>
We will employ various preprocessing tasks such as:<br>

•  ColumnTransformer to scale the year, tempo and duration (i.e. convert to 0-1 scale)<br>
•  One-hot encoding (i.e create 10 columns taking 0/1 values)

And because 'artists' and 'name' cannot be converted to numerical values, we will drop them for now.


```python
ct = ColumnTransformer([('minmax', MinMaxScaler(), ['year', 'tempo', 'duration_ms']),
                        ('categorical', OneHotEncoder(), ['key']),
                        ('drop_cols', 'drop', ['artists', 'name'])],
                       remainder='passthrough')

ct.fit(X_train)

X_train_preprocessed = ct.transform(X_train)
X_val_preprocessed = ct.transform(X_val)
X_test_preprocessed = ct.transform(X_test)
```

```python
X_train_preprocessed.shape
```

We will now check the Mean Square Error (MSE) with various regresssion models

```python
# Linear Regression

lin_reg = LinearRegression()
lin_reg.fit(X_train_preprocessed, y_train)
y_pred = lin_reg.predict(X_val_preprocessed)

mean_squared_error(y_val, y_pred, squared=False)
```

```python
# Lasso

lasso = Lasso()
lasso.fit(X_train_preprocessed,y_train)
y_pred = lasso.predict(X_val_preprocessed)

mean_squared_error(y_val, y_pred, squared=False)
```

```python
# K-Nearest-Neighbors (KNN) regressor

neigh = KNeighborsRegressor(n_neighbors=7)
neigh.fit(X_train_preprocessed, y_train)
y_pred = neigh.predict(X_val_preprocessed)

mean_squared_error(y_val, y_pred, squared=False)
```

<font color='red'>FINDING: From the above tests, KNN has the lowest Mean Square Error</font>

<h2>4.5 Testing the Model</h2>
<br>
Now we will train our model on a larger train set, composed of the former `X_train` and `X_val`, and test its performances on `X_test`. <br>

```python
X_train_old = X_train
X_train = pd.concat([X_train_old, X_val])
```

```python
y_train_old = y_train.copy()
y_train = pd.concat([y_train_old, y_val])
```

```python
# Defining a new dataframe that corresponds to X_train, but where popularity has not been dropped. 
song_train = song.sort_values(by='year').loc[:120014].copy() 
```

```python
# At the plot 'Mean Popularity Over The Years', there is a sudden drop of popularity around the end of 2020.
# This may be due to Spotify's data not updating to the latest metrics.
# And since the popularity scores tend to lag on the Spotify API, we drop the songs from 2021.

X_test_no_2021=X_test[X_test['year']!=2021].copy()
y_test_no_2021=y_test[:X_test_no_2021.shape[0]].copy() 
```

```python
# Preprocessing the 'artists' column as seen before

artists_and_pop = {}

train_mean_pop = song_train['popularity'].mean()

for artist in X_train['artists'].unique():
    temp = song_train[song_train['artists'] == artist]['popularity'].copy()
    if len(temp) > 1:
        artists_and_pop[artist] = temp.mean()
    elif len(temp) == 1:
        artists_and_pop[artist] = train_mean_pop

X_train['artists'] = X_train['artists'].map(artists_and_pop)
X_test_no_2021['artists'] = X_test_no_2021['artists'].map(lambda artist: artists_and_pop.get(artist, train_mean_pop))
```

```python
# Adding usual preprocessing steps

ct = ColumnTransformer([('minmax', MinMaxScaler(), ['year', 'tempo', 'duration_ms', 'artists']),
                        ('categorial', OneHotEncoder(), ['key']),
                        ('drop_cols', 'drop', ['name'])],
                       remainder='passthrough')

ct.fit(X_train)

X_train_preprocessed = ct.transform(X_train)
X_test_preprocessed = ct.transform(X_test_no_2021)
```

<font color='red'>SCORING THE KNN REGRESSOR
<br>
Now that we have reconstituted our new train set, we can score our best model (KNN with 18 neighbors) on the test set.<font/>


```python
neigh = KNeighborsRegressor(n_neighbors=18)
neigh.fit(X_train_preprocessed, y_train)
y_pred = neigh.predict(X_test_preprocessed)

mean_squared_error(y_test_no_2021, y_pred, squared=False)
```

This is our **mean squared error on the test set**, ie. songs issued roughly between 2014 and 2020 (10% of the whole dataset).
